```python
import pandas as pd
file_path = '/content/drive/MyDrive/Datasets-20241209T022939Z-001/Datasets/medical_students_dataset.csv'
medical_df = pd.read_csv(file_path)
print("Dataset Loaded Successfully:")
medical_df.head()
```

```python
print("DataFrame Information:")
medical_df.info()
null_values = medical_df.isnull().sum()
print("Columns with missing values:")
print(null_values[null_values > 0])
```

For categorical columns, missing values are filled with the mode (most frequent value) to maintain consistency. For numeric columns, missing values are filled with the mean to preserve the data's overall distribution.

```python
import pandas as pd
df = pd.read_csv('/content/drive/MyDrive/Datasets-20241209T022939Z-001/Datasets/medical_students_dataset.csv')
for column in df.columns:
    if df[column].dtype == 'object':
        df[column].fillna(df[column].mode()[0], inplace=True)
    else:
        df[column].fillna(df[column].mean(), inplace=True)
print(df.head())
```

```python
duplicates_count = medical_df.duplicated().sum()
print(f"Number of duplicate rows in the dataset: {duplicates_count}")

if duplicates_count > 0:
    medical_df = medical_df.drop_duplicates()
    print("Duplicate rows have been removed.")
print(f"Number of duplicate rows after removal: {medical_df.duplicated().sum()}")
```

```python
import pandas as pd
file_path = '/content/drive/MyDrive/Datasets-20241209T022939Z-001/Datasets/Titanic-Dataset.csv'
titanic_df = pd.read_csv(file_path)
subset_df = titanic_df[['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']]
first_class_df = subset_df[subset_df['Pclass'] == 1]
fare_mean = first_class_df['Fare'].mean()
fare_median = first_class_df['Fare'].median()
fare_max = first_class_df['Fare'].max()
fare_min = first_class_df['Fare'].min()

print(f"Fare - Mean: {fare_mean}, Median: {fare_median}, Max: {fare_max}, Min: {fare_min}")
```

```python
null_age_count = first_class_df['Age'].isnull().sum()
print(f"Null values in 'Age': {null_age_count}")
first_class_df = first_class_df.dropna(subset=['Age'])
print(f"Null values in 'Age' after drop: {first_class_df['Age'].isnull().sum()}")
```

```python
embarked = pd.get_dummies(titanic_df['Embarked'], prefix='Embarked')
titanic_df = pd.concat([titanic_df, embarked], axis=1)
titanic_df = titanic_df.drop('Embarked', axis=1)
print(titanic_df.head())
```

```python
import matplotlib.pyplot as plt
mean_survival_by_gender = titanic_df.groupby('Sex')['Survived'].mean()

mean_survival_by_gender.plot(kind='bar', color=['black', 'grey'])
plt.title('Mean Survival Rate by Gender')
plt.ylabel('Mean Survival Rate')
plt.xticks(rotation=0)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
mean_survival_by_embarked = {
    'Cherbourg': titanic_df[titanic_df['Embarked_C'] == 1]['Survived'].mean(),
    'Queenstown': titanic_df[titanic_df['Embarked_Q'] == 1]['Survived'].mean(),
    'Southampton': titanic_df[titanic_df['Embarked_S'] == 1]['Survived'].mean()
}
embarked_ports = list(mean_survival_by_embarked.keys())
survival_rates = list(mean_survival_by_embarked.values())

plt.figure(figsize=(6, 4))
sns.barplot(x=embarked_ports, y=survival_rates, palette='viridis')
plt.title('Mean Survival Rate by Port of Embarkation')
plt.xlabel('Port of Embarkation')
plt.ylabel('Mean Survival Rate')
plt.show()
```
